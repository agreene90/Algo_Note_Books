# Viewing the Past: The Relationship Between Light, Distance, and Time

## Introduction
This notebook explores the concept of viewing the past through the finite speed of light. By understanding how light travels over time and distance, we can calculate how far into the past we are seeing an object based on its distance from us.

## The Physics of Light and Time
The key equation is \( t = rac{d}{c} \), where:
- \( t \) is the time into the past we are observing.
- \( d \) is the distance to the object.
- \( c \) is the speed of light, approximately 299,792,458 meters per second.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

## Function to Calculate Time Past

```python
def calculate_time_past(distance):
    c = 299792458 # Speed of light in meters per second
    return distance / c
```

## Visualizing Time Past for Different Distances

```python
def plot_time_past(distances):
    time_delays = calculate_time_past(distances)
    plt.figure(figsize=(10, 6))
    plt.plot(distances, time_delays, color='blue')
    plt.title('Time into the Past vs Distance')
    plt.xlabel('Distance (meters)')
    plt.ylabel('Time into the Past (seconds)')
    plt.grid(True)
    plt.show()

initial_distances = np.linspace(1, 1e9, 1000)
plot_time_past(initial_distances)
```

## Interactive GUI for Custom Distance Input

```python
@widgets.interact_manual(distance=widgets.FloatLogSlider(value=10, base=10, min=0, max=9, step=0.1, description='Distance (m)'))
def interactive_view(distance):
    time_past = calculate_time_past(distance)
    print(f"Time into the past: {time_past:.10f} seconds")
```

## Conclusion and Further Exploration
This notebook provides a basic framework for exploring the relationship between distance, the speed of light, and observing the past. By adjusting the distance, we can see how it affects our 'view' into the past, albeit on a small scale compared to astronomical observations.
