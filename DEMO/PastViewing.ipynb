# Viewing the Past: The Relationship Between Light, Distance, and Time

## Introduction
This notebook explores the concept of viewing the past through the finite speed of light. By understanding how light travels over time and distance, we can calculate how far into the past we are seeing an object based on its distance from us.

## The Physics of Light and Time
The key equation is \( t = rac{d}{c} \), where:
- \( t \) is the time into the past we are observing.
- \( d \) is the distance to the object.
- \( c \) is the speed of light, approximately 299,792,458 meters per second.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

## Function to Calculate Time Past

```python
def distance_for_time(time_minutes):
    speed_of_light = 299792458 # Speed of light in meters per second
    time_seconds = time_minutes * 60
    return speed_of_light * time_seconds
```

## Visualizing Time Past for Different Distances

```python
def calculate_time_past(distance):
    speed_of_light = 299792458 # Speed of light in meters per second
    if distance < 0:
        raise ValueError("Distance cannot be negative.")
    return distance / speed_of_light

def plot_time_past(distances, time_mark):
    time_delays = [calculate_time_past(d) for d in distances]
    plt.figure(figsize=(10, 6))
    plt.plot(distances, time_delays, color='blue')
    plt.axhline(y=time_mark * 60, color='r', linestyle='--')
    plt.title('Time into the Past vs Distance')
    plt.xlabel('Distance (meters)')
    plt.ylabel('Time into the Past (seconds)')
    plt.grid(True)
    plt.show()

time_mark = 20 # 20 minutes
max_distance = distance_for_time(time_mark)
initial_distances = np.linspace(1, max_distance, 1000)
plot_time_past(initial_distances, time_mark)
```

## Interactive GUI for Custom Distance Input

```python
max_time_minutes = 20  # Scaling up to 20 minutes
max_distance = distance_for_time(max_time_minutes)

distance_slider = widgets.FloatLogSlider(
    value=10,
    base=10,
    min=0, 
    max=np.log10(max_distance),
    step=0.1,
    description='Distance (m)'
)

def interactive_view(distance):
    time_past = calculate_time_past(distance)
    print(f"Time into the past: {time_past / 60:.10f} minutes")

widgets.interact_manual(interactive_view, distance=distance_slider)
```

## Conclusion and Further Exploration
